```python
#area plot in data visualization
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Product A': [100, 150, 200, 250, 300],
    'Product B': [80, 120, 160, 190, 230]
}

df = pd.DataFrame(data)

# Set 'Year' as the index for plotting
df.set_index('Year', inplace=True)

# Area plot
df.plot(kind='area', stacked=True, alpha=0.6)

# Title and labels
plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.show()

```

```python
#data visualization using histrogram
#matplotlib
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = [25, 30, 35, 40, 40, 45, 50, 50, 50, 60, 70, 75, 80, 90, 95]

# Create histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')

# Labels and title
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
#using seaborn lib
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (can also use a column from a DataFrame)
data = [25, 30, 35, 40, 40, 45, 50, 50, 50, 60, 70, 75, 80, 90, 95]

# Create histogram with Seaborn
sns.histplot(data, bins=10, kde=False, color='green')

# Title and labels
plt.title('Seaborn Histogram Example')
plt.xlabel('Value')
plt.ylabel('Count')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

# Create bar chart
plt.bar(categories, values, color='skyblue')

# Add title and labels
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value': [10, 24, 36, 18]
}
df = pd.DataFrame(data)

# Create bar chart with Seaborn
sns.barplot(x='Category', y='Value', data=df, palette='pastel')

# Add title
plt.title('Seaborn Bar Chart Example')

plt.show()
```
