```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from URL
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-car-sales.csv'
df = pd.read_csv(url)

# Extract Year from 'Month' column (which is in format YYYY-MM)
df['Year'] = pd.to_datetime(df['Month']).dt.year

# Group by Year and sum sales
yearly_sales = df.groupby('Year')['Sales'].sum()

# Plot line chart
yearly_sales.plot(kind='line', marker='o', figsize=(10, 6), title='Yearly Automobile Sales')
plt.xlabel('Year')
plt.ylabel('Total Automobile Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Filter recession and non-recession data
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

# Create figure and subplots
fig = plt.figure(figsize=(14, 6))

# Subplot 1: Recession Period
ax0 = fig.add_subplot(1, 2, 1)
sns.lineplot(x='Year', y='GDP', data=rec_data, marker='o', ax=ax0)
ax0.set_title('GDP during Recession Period')
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')

# Subplot 2: Non-Recession Period
ax1 = fig.add_subplot(1, 2, 2)
sns.lineplot(x='Year', y='GDP', data=non_rec_data, marker='o', ax=ax1)
ax1.set_title('GDP during Non-Recession Period')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Filter the data
Rdata = df[df['Recession'] == 1]
NRdata = df[df['Recession'] == 0]

# Calculate the total advertising expenditure for both periods
RAtotal = Rdata['Advertising_Expenditure'].sum()
NRAtotal = NRdata['Advertising_Expenditure'].sum()

# Create a pie chart for the advertising expenditure
plt.figure(figsize=(8, 6))

labels = ['Recession', 'Non-Recession']
sizes = [RAtotal, NRAtotal]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Advertising Expenditure: Recession vs Non-Recession Periods')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Filter data for recession period
Rdata = df[df['Recession'] == 1]

# Calculate total sales volume by vehicle type during recession
VTexpenditure = Rdata.groupby('Vehicle_Type')['Automobile_Sales'].sum()

# Create a pie chart for vehicle type share during recession
plt.figure(figsize=(8, 6))

labels = VTexpenditure.index
sizes = VTexpenditure.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Vehicle Type Share of Automobile Sales During Recession')
plt.axis('equal')  # Ensures pie chart is circular
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Filter data for recession period
df_rec = df[df['Recession'] == 1]

# Plot lineplot for sales trend during recession by vehicle type
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=df_rec,
    x='Year',
    y='Automobile_Sales',
    hue='Vehicle_Type',
    style='Vehicle_Type',
    markers=True,
    dashes=False,
    err_style=None
)

plt.ylim(0, 850)
plt.title('Vehicle Sales Trend by Type During Recession')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.legend(loc=(0.05, 0.3))
plt.grid(True)
plt.tight_layout()
plt.show()
```
